```sage
from sage.plot.plot3d.shapes import *
s = sphere(opacity=0.5,aspect_ratio=[1,1,1])
#s = Sphere(1.0)
box_size=2
b = Box([box_size,box_size,0.01],color="gray",opacity=0.5)
```

```sage
u=1.0
v=1.5
p1 = point((u,v,0),color="red",size=20)
```

```sage
d = u*u+v*v+1
p2 = point((2*u/d,2*v/d,(u*u+v*v-1)/d),color="red",size=20)
```

```sage
l = line([p1.loc,p2.loc],color="black")
```

```sage
s+b+p1+p2+l
```

```sage
g = s+b
du=0.4
dv=0.4
u0=1.0
v0=1.5
for i in range(2):
    for j in range(2):
        u=u0+du*i
        v=v0+dv*j
        p1 = point((u,v,0.05),color="white",size=20)
        d = u*u+v*v+1
        p2 = point((2*u/d,2*v/d,(u*u+v*v-1)/d+0.05),color="white",size=20)
        l = line([p1.loc,p2.loc],color="red")
        g += p1+p2+l
g
```

```sage
u,v=var('u,v')
n=u*u+v*v
pp = parametric_plot((2*u/(n+1),2*v/(n+1),(n-1)/(n+1)),(u,u0,u0+du),(v,v0,v0+dv),color="green")
g+pp
```
